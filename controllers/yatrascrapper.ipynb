```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import warnings
import csv
warnings.filterwarnings('ignore')
```

```python
ataglimse = []
statetourism = []
besttime = []
touristatt = []
howtoreach = []
gettingaround = []
curr = []
popu = []
timezone = []
ar = []
cap = []
officiallan = []
callcode = []
reg = []
citynames = []
statenames = []
rl = []
row = []
```

```python
listofstates = ['https://www.yatra.com/india-tourism/haryana', 'https://www.yatra.com/india-tourism/himachal-pradesh', 'https://www.yatra.com/india-tourism/jammu-and-kashmir', 'https://www.yatra.com/india-tourism/punjab', 'https://www.yatra.com/india-tourism/rajasthan', 'https://www.yatra.com/india-tourism/uttar-pradesh', 'https://www.yatra.com/india-tourism/uttarakhand', 'https://www.yatra.com/india-tourism/delhi-ncr', 'https://www.yatra.com/india-tourism/andhra-pradesh', 'https://www.yatra.com/india-tourism/telangana', 'https://www.yatra.com/india-tourism/karnataka', 'https://www.yatra.com/india-tourism/kerala', 'https://www.yatra.com/india-tourism/tamil-nadu', 'https://www.yatra.com/india-tourism/andaman', 'https://www.yatra.com/india-tourism/lakshadweep-islands', 'https://www.yatra.com/india-tourism/goa', 'https://www.yatra.com/india-tourism/gujarat', 'https://www.yatra.com/india-tourism/maharashtra', 'https://www.yatra.com/india-tourism/dadra-and-nager-haveli', 'https://www.yatra.com/india-tourism/daman-and-diu', 'https://www.yatra.com/india-tourism/bihar', 'https://www.yatra.com/india-tourism/jharkhand', 'https://www.yatra.com/india-tourism/orissa', 'https://www.yatra.com/india-tourism/west-bengal', 'https://www.yatra.com/india-tourism/arunachal-pradesh', 'https://www.yatra.com/india-tourism/assam', 'https://www.yatra.com/india-tourism/manipur', 'https://www.yatra.com/india-tourism/meghalaya', 'https://www.yatra.com/india-tourism/mizoram', 'https://www.yatra.com/india-tourism/nagaland', 'https://www.yatra.com/india-tourism/sikkim', 'https://www.yatra.com/india-tourism/tripura', 'https://www.yatra.com/india-tourism/chhattisgarh', 'https://www.yatra.com/india-tourism/madhya-pradesh']
for x in listofstates:
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(x)
    at_a_glimpse = driver.find_element_by_css_selector('#travel_guide p').text
    ataglimse.append(at_a_glimpse)
    rl.append(at_a_glimpse)
    btn = driver.find_element_by_css_selector('.cg-state-tabmsc_1')
    btn.click()
    state_tourism = driver.find_element_by_css_selector('#tourism_in p').text
    statetourism.append(state_tourism)
    rl.append(state_tourism)
    btn = driver.find_element_by_css_selector('.cg-state-tabmsc_2')
    btn.click()
    best_time = driver.find_element_by_css_selector('#best_time_to_visit p').text
    besttime.append(best_time)
    rl.append(best_time)
    btn = driver.find_element_by_css_selector('.cg-state-tabmsc_3')
    btn.click()
    tourist_att = driver.find_element_by_css_selector('#tourist_attractions p').text
    touristatt.append(tourist_att)
    rl.append(tourist_att)
    btn = driver.find_element_by_css_selector('.cg-state-tabmsc_4')
    btn.click()
    how_to_reach = driver.find_element_by_css_selector('#how_to_reach p').text
    howtoreach.append(how_to_reach)
    rl.append(how_to_reach)
    btn = driver.find_element_by_css_selector('.cg-state-tabmsc_5')
    btn.click()
    getting_around = driver.find_element_by_css_selector('#travel_in p').text
    gettingaround.append(getting_around)
    rl.append(getting_around)
    currency = driver.find_element_by_css_selector('li:nth-child(1) .c435061').text
    population = driver.find_element_by_css_selector('li:nth-child(2) .c435061').text
    time_zone = driver.find_element_by_css_selector('li:nth-child(3) .c435061').text
    area = driver.find_element_by_css_selector('li:nth-child(4) .c435061').text
    capital = driver.find_element_by_css_selector('li:nth-child(5) .c435061').text
    official_language = driver.find_element_by_css_selector('li:nth-child(6) .c435061').text
    calling_code = driver.find_element_by_css_selector('li:nth-child(7) .c435061').text
    try:
        regions = driver.find_element_by_css_selector('.regions').text
        ls = regions.split(' ')[2:]
        if(len(ls)!= 1):
            regions = ' '.join(ls)
    except:
        print(state_name)
        regions = "None here"
    city_names = regions.split('\n')[1:]
    state_name = regions.split('\n')[0]
    curr.append(currency)
    rl.append(currency)
    popu.append(population)
    rl.append(population)
    timezone.append(time_zone)
    rl.append(time_zone)
    ar.append(area)
    rl.append(area)
    cap.append(capital)
    rl.append(capital)
    officiallan.append(official_language)
    rl.append(official_language)
    callcode.append(calling_code)
    rl.append(calling_code)
    reg.append(regions)
    rl.append(regions)
    citynames.append(city_names)
    rl.append(city_names)
    statenames.append(state_name)
    rl.append(state_name)
    row.append(rl)
    rl = []
    
    # state_name
    # city_names
   
    driver.close()
    time.sleep(2)
```
